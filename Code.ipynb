# K-Mean Clsutering with 10 Clusters

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def load_breast_cancer_dataset():
    # Load the Breast Cancer dataset
    data = load_breast_cancer()
    features = data.data
    labels = data.target
    feature_names = data.feature_names
    target_names = data.target_names
    return features, labels, feature_names, target_names

if __name__ == "__main__":
    features, labels, feature_names, target_names = load_breast_cancer_dataset()

    # Apply PCA to reduce the number of features to 2 principal components
    pca = PCA(n_components=2)
    principal_components = pca.fit_transform(features)

    # Implement K-means clustering with 10 clusters
    kmeans = KMeans(n_clusters=10, random_state=42)
    clusters = kmeans.fit_predict(principal_components)

    # Plot all observations on a 2D coordinate system with colored clusters
    plt.figure(figsize=(10, 8))

    unique_clusters = np.unique(clusters)
    colors = plt.cm.tab10(np.linspace(0, 1, len(unique_clusters)))

    for i, cluster_label in enumerate(unique_clusters):
        cluster_mask = clusters == cluster_label
        plt.scatter(
            principal_components[cluster_mask, 0],
            principal_components[cluster_mask, 1],
            color=colors[i],
            label=f'Cluster {cluster_label}'
        )

    plt.title("K-means Clustering with 10 Clusters")
    plt.xlabel("Principal Component 1")
    plt.ylabel("Principal Component 2")
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Use hierarchical clustering with 5 clusters (use AgglomerativeClustering(n_clusters = 5) instead of KMeans(n_clusters = 10))

# Hierarchical clustering with 20 clusters (Agglomerative Clustering) for the digits dataset

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

def load_digits_dataset():
    # Load the digits dataset
    data = load_digits()
    features = data.data
    labels = data.target
    return features, labels

if __name__ == "__main__":
    features, labels = load_digits_dataset()

    # Apply PCA to reduce the number of features to 2 principal components
    pca = PCA(n_components=2)
    principal_components = pca.fit_transform(features)

    # Implement Agglomerative Clustering with 20 clusters
    agglomerative = AgglomerativeClustering(n_clusters=20)
    clusters = agglomerative.fit_predict(principal_components)

    # Create a DataFrame with the principal components and cluster labels
    df = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])
    df['cluster'] = clusters

    # Plot observations for each cluster separately
    plt.figure(figsize=(10, 8))

    unique_clusters = np.unique(clusters)
    colors = plt.cm.tab20(np.linspace(0, 1, len(unique_clusters)))

    for i, cluster_label in enumerate(unique_clusters):
        cluster_mask = clusters == cluster_label
        plt.scatter(
            principal_components[cluster_mask, 0],
            principal_components[cluster_mask, 1],
            color=colors[i],
            label=f'Cluster {cluster_label}'
        )

    plt.title("Agglomerative Clustering with 20 Clusters (PCA)")
    plt.xlabel("Principal Component 1")
    plt.ylabel("Principal Component 2")
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Comparing the results

```python
from sklearn.metrics import silhouette_score

def evaluate_clustering(principal_components, labels, clusters):
    # Calculate the silhouette score
    silhouette = silhouette_score(principal_components, clusters)

    # If the labels are available, calculate the inertia for K-means
    if labels is not None:
        from sklearn.metrics import mean_squared_error
        kmeans_inertia = mean_squared_error(labels, clusters)

        return silhouette, kmeans_inertia

    return silhouette, None

# Load datasets and perform clustering
breast_cancer_features, breast_cancer_labels, _, _ = load_breast_cancer_dataset()
digits_features, digits_labels = load_digits_dataset()
iris_features, iris_labels, _ = load_iris_dataset()

pca = PCA(n_components=2)

# K-means clustering
pca_breast_cancer = pca.fit_transform(breast_cancer_features)
kmeans = KMeans(n_clusters=10, random_state=42)
kmeans_clusters = kmeans.fit_predict(pca_breast_cancer)

# DBSCAN clustering
pca_digits = pca.fit_transform(digits_features)
dbscan = DBSCAN(min_samples=10, eps=1.5)
dbscan_clusters = dbscan.fit_predict(pca_digits)

# Agglomerative Clustering
pca_iris = pca.fit_transform(iris_features)
agglomerative = AgglomerativeClustering(n_clusters=5)
agglomerative_clusters = agglomerative.fit_predict(pca_iris)

# Evaluate the clustering results
silhouette_kmeans, inertia_kmeans = evaluate_clustering(pca_breast_cancer, breast_cancer_labels, kmeans_clusters)
silhouette_dbscan, _ = evaluate_clustering(pca_digits, digits_labels, dbscan_clusters)
silhouette_agglomerative, _ = evaluate_clustering(pca_iris, iris_labels, agglomerative_clusters)

# Print the evaluation results
print("K-means Silhouette Score:", silhouette_kmeans)
print("K-means Inertia:", inertia_kmeans)
print("DBSCAN Silhouette Score:", silhouette_dbscan)
print("Agglomerative Clustering Silhouette Score:", silhouette_agglomerative)
```

```python
import matplotlib.pyplot as plt

# Data for plotting
methods = ['K-means', 'DBSCAN', 'Agglomerative']
silhouette_scores = [silhouette_kmeans, silhouette_dbscan, silhouette_agglomerative]

# Plot the silhouette scores for each clustering method
plt.figure(figsize=(8, 6))
plt.bar(methods, silhouette_scores, color=['blue', 'green', 'orange'])
plt.title("Silhouette Scores for Clustering Methods")
plt.ylabel("Silhouette Score")
plt.ylim(-0.5, 0.6)  # Adjust the y-axis limits for better visualization
plt.grid(True)
plt.show()
```
